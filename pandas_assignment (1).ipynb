## Q1. List any five functions of the pandas library with execution.

**Answer:** Below are five pandas functions with examples.

```python
import pandas as pd

data = {'A':[1,2,3], 'B':[4,5,6]}
df = pd.DataFrame(data)

# 1. head()
print("head():")
print(df.head())

# 2. describe()
print("\ndescribe():")
print(df.describe())

# 3. info()
print("\ninfo():")
df.info()

# 4. sum()
print("\nsum():")
print(df.sum())

# 5. mean()
print("\nmean():")
print(df.mean())
```

## Q2. Re-index the DataFrame starting from 1 and incrementing by 2.

```python
import pandas as pd

df = pd.DataFrame({
    'A':[10,20,30],
    'B':[40,50,60],
    'C':[70,80,90]
})

df.index = range(1, 2*len(df)+1, 2)
print(df)
```

## Q3. Calculate the sum of the first three values in the 'Values' column.

```python
import pandas as pd

df = pd.DataFrame({'Values':[10,20,30,40,50]})

def sum_first_three(df):
    total = 0
    for i in range(3):
        total += df['Values'][i]
    print("Sum of first three values:", total)

sum_first_three(df)
```

## Q4. Create a new column 'Word_Count' that counts the number of words in each row of the 'Text' column.

```python
import pandas as pd

df = pd.DataFrame({
    'Text':["Hello world",
            "Pandas is powerful",
            "Data science is interesting"]
})

df['Word_Count'] = df['Text'].apply(lambda x: len(x.split()))
print(df)
```

## Q5. Difference between DataFrame.size() and DataFrame.shape()

**Answer:**

- **DataFrame.size** → Returns the total number of elements in the DataFrame (rows × columns).
- **DataFrame.shape** → Returns a tuple representing (rows, columns).


```python
import pandas as pd

df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})

print("Size:", df.size)
print("Shape:", df.shape)
```

## Q7. Extract username from email and create a new column 'Username'.

```python
import pandas as pd

df = pd.DataFrame({
    'Email':['john.doe@example.com',
             'alice@gmail.com',
             'bob@yahoo.com']
})

df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])

print(df)
```

## Q8. Select rows where A > 5 and B < 10.

```python
import pandas as pd

df = pd.DataFrame({
    'A':[3,8,6,2,9],
    'B':[5,2,9,3,1],
    'C':[1,7,4,5,2]
})

filtered_df = df[(df['A'] > 5) & (df['B'] < 10)]

print(filtered_df)
```
